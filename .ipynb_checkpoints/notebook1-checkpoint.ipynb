```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
%matplotlib inline
```

```python
df = pd.read_csv('eye_movements.csv')
df.head
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

```python
y = df['label']
X = df.drop(columns=['label'])
X.head
```

```python
X.shape
```

```python
y.head
```

```python
from sklearn.model_selection import train_test_split

X_a, X_test, y_a, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
X_train, X_valid, y_train, y_valid = train_test_split(X_a, y_a, test_size=0.2, random_state=51)
```

```python
clf = RandomForestClassifier(max_depth=4, random_state=0)
```

```python
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn import model_selection
```

```python
def fit_ml_algo(algo, X_train, y_train, cv):
  model = algo.fit(X_train, y_train)
  acc = round(model.score(X_train, y_train) * 100, 2)
  train_pred = model_selection.cross_val_predict(algo, X_valid, y_valid, cv=cv, n_jobs=-1)
  acc_cv = round(sklearn.metrics.accuracy_score(y_valid, train_pred) * 100, 2)
  return train_pred, acc, acc_cv
```

```python
_, acc, acc_cv = fit_ml_algo(AdaBoostClassifier(), X_train, y_train, 5)
acc, acc_cv
```

```python
for depth in range(1, 100):
  _, acc, acc_cv = fit_ml_algo(RandomForestClassifier(max_depth=depth), X_train, y_train, 2)
  print("Depth:", depth, "Metrics:", acc, acc_cv)
```

```python
acc_cv
```

```python
from sklearn import neighbors
for n_neighbors in range(1, 25):
  for weights in ['uniform', 'distance']:
    _, acc, acc_cv = fit_ml_algo(neighbors.KNeighborsClassifier(n_neighbors, weights=weights), X_train, y_train, 5)
    print(weights, n_neighbors, acc, acc_cv)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers

y_new = tf.keras.utils.to_categorical(y, num_classes=3)

model = tf.keras.Sequential([
                             tf.keras.layers.Dense(27, activation='relu'),
                             tf.keras.layers.Dense(16, activation='relu'),
                             tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
model.compile(keras.optimizers.Adam(0.01), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y_new, epochs=300, batch_size=1)
```
